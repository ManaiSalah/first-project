# Importing the libraries we need to work with and importing my file using Pandas.

```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
movies=pd.read_csv("/kaggle/input/imdb-dataset-of-top-1000-movies-and-tv-shows/imdb_top_1000.csv",index_col=0)
```

# Informations about our Data 

```python
movies.head()  

movies.columns

movies.describe

movies.dtypes


```

# Getting rid of Null and missing values

```python
missing_values= movies.isnull().sum() #number of null values in our data 

movies.dropna(inplace=True)

movies.isnull().sum()



```

# The relationship between the gross and the ratings 

```python
gross= movies["Gross"]
gross_count=gross.count()
gross_mean=gross.mean



rate=movies["IMDB_Rating"]
rate.max()        #the highest rating
rate.min()        #the lowest rating 
rate.mean()       #the mean rating 


plt.scatter(gross,rate)
plt.title("the relationship between gross and ratings")
plt.ylabel("ratings")
plt.xlabel("gross")
plt.show()


```
